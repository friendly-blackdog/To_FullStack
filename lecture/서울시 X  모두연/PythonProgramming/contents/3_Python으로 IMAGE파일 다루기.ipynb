Imagefile
- 디지털 사진은 수많은 점(픽셀)들로 구성되어 있습니다.
- 수많은 점들을 자동으로 읽고 쓰기 위해 반복문이 필요합니다.
- 각각의 점은 Color의 RGB값을 나타내며, 이는 Tuple로 표현합니다.
- Color는 투명도를 포함하는 RGBA 4개의 len의 Tuple로 표현하기도한다
- Imagefile의 좌표는 왼쪽 위를 0,0을 시작으로 오른쪽, 아래로 내려갈수로 X방향(오른쪽) // Y방향(아래쪽)으로 좌표가 증가한다.

## 3) 흑백사진으로 변환하기

흑백사진은 RGB값에 따라 아래와 같은 밝기로 나태낼 수 있다. </br> 
L = 0.213 R + 0.715 G + 0.072 B

```
def luma(p):
    r, g, b = p
    return int(0.213 * r + 0.715 * g + 0.072 * b)
```

````python
creator_jang = Image.open("Creator_Jang.png")
pix = creator_jang.load()
w, h = creator_jang.size
for x in range(w):
    for y in range(h):
        L=0.213*pix[x,y][0]+0.715*pix[x,y][1]+0.072*pix[x,y][2]
        L=int(L)
        pix[x,y]=(L,L,L)
creator_jang
````
