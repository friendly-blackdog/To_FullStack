```python
from tkinter import *
import tkinter.messagebox

def read_depots(file):
    depots = []
    depots_f = open(file)
    for line in depots_f:
        depots.append(line.rstrip())
    return depots

def save_data():
    try:
        fileD = open("deliveries.txt", "a")
        fileD.write("Depot:\n")
        fileD.write("%s\n" % depot.get())
        fileD.write("Description:\n")
        fileD.write("%s\n" % description.get())
        fileD.write("Address:\n")
        fileD.write("%s\n" % address.get("1.0", END))
        depot.set("")
        description.delete(0, END)
        address.delete("1.0", END)
    except Exception as ex:
        tkinter.messagebox.showinfo("Error!", "Can't write to the file\n %s" % ex)

app = Tk()
app.title('Head-Ex Deliveries')

Label(app, text = "Depot:").pack()
depot = StringVar()
depot.set(None)
options = read_depots("depots.txt")
OptionMenu(app, depot, *options).pack()

Label(app, text = "Description:").pack()
description = Entry(app)
description.pack()

Label(app, text = "Address:").pack()
address = Text(app)
address.pack()

Button(app, text = "Save", command = save_data).pack()

app.mainloop()
```

```python
from tkinter import *
from tkinter import messagebox
from db import Database

db = Database('sales.db')


def populate_list():
    parts_list.delete(0, END)
    for row in db.fetch():
        parts_list.insert(END, row)

def add_item():
    if part_text.get() == '' or maker_text.get() == '' or quantity_text.get() == '' or price_text.get() == '':
        messagebox.showerror('Required Fields', 'Plese include all fields')
        return
    db.insert(part_text.get(), maker_text.get(), quantity_text.get(), price_text.get())
    parts_list.delete(0, END)
    parts_list.insert(END, (part_text.get(), maker_text.get(), quantity_text.get(),
                            price_text.get()))
    populate_list()

def select_item(event):
    try:
        global selected_item
        index = parts_list.curselection()[0]
        selected_item = parts_list.get(index)

        part_entry.delete(0, END)
        part_entry.insert(END, selected_item[1])
        maker_entry.delete(0, END)
        maker_entry.insert(END, selected_item[2])
        quantity_entry.delete(0, END)
        quantity_entry.insert(END, selected_item[3])
        price_entry.delete(0, END)
        price_entry.insert(END, selected_item[4])
    except IndexError:
        pass

def remove_item():
    db.remove(selected_item[0])
    clear_item()
    populate_list()

def update_item():
    db.update(selected_item[0], part_text.get(), maker_text.get(), quantity_text.get(), price_text.get())
    populate_list()


def clear_item():
    part_entry.delete(0, END)
    maker_entry.delete(0, END)
    quantity_entry.delete(0, END)
    price_entry.delete(0, END)


# Create window object
window = Tk()

#Part
part_text = StringVar()
part_label = Label(window, text='Part Name', font=('bold', 14), pady=20)
part_label.grid(row=0, column=0)
part_entry = Entry(window, textvariable=part_text)
part_entry.grid(row=0, column=1)

#Maker
maker_text = StringVar()
maker_label = Label(window, text='maker', font=('bold', 14), pady=20)
maker_label.grid(row=0, column=2)
maker_entry = Entry(window, textvariable=maker_text)
maker_entry.grid(row=0, column=3)

#quantity
quantity_text = StringVar()
quantity_label = Label(window, text='quantity', font=('bold', 14), pady=20)
quantity_label.grid(row=1, column=0)
quantity_entry = Entry(window, textvariable=quantity_text)
quantity_entry.grid(row=1, column=1)

#Price
price_text = StringVar()
price_label = Label(window, text='price', font=('bold', 14), pady=20)
price_label.grid(row=1, column=2)
price_entry = Entry(window, textvariable=price_text)
price_entry.grid(row=1, column=3)

#Parts lists (Listbox)
parts_list = Listbox(window, height=8, width=50, border=0)
parts_list.grid(row=3, column=0, columnspan=3, rowspan=6, pady=20, padx=20)

#Create scrollbar
scrollbar = Scrollbar(window)
scrollbar.grid(row=3, column=3)

#Set scroll to listbox
parts_list.configure(yscrollcommand=scrollbar.set)
scrollbar.configure(command=parts_list.yview)

# Bind select
parts_list.bind('<<ListboxSelect>>', select_item)

#Buttons
add_btn = Button(window, text='Add Part', width=12, command=add_item)
add_btn.grid(row=2, column=0, pady=20, )

remove_btn = Button(window, text='Remove Part', width=12, command=remove_item)
remove_btn.grid(row=2, column=1, pady=20, )

update_btn = Button(window, text='Update Part', width=12, command=update_item)
update_btn.grid(row=2, column=2, pady=20, )

clear_btn = Button(window, text='Clear Input', width=12, command=clear_item)
clear_btn.grid(row=2, column=3, pady=20, )


window.title('Sales Manager')
window.geometry('700x350')

window.mainloop()
```

```python
from tkinter import *
import tkinter.ttk as ttk

window = Tk()
window.title("ZOO")


animal = ttk.Entry(window, width=30).grid(row=0, column=1, padx=20)
number = ttk.Entry(window, width=30).grid(row=1, column=1, padx=20)
price = ttk.Entry(window, width=30).grid(row=2, column=1, padx=20)


def update():
    conn = sqlite3.connect('zoo.db')
    c = conn.cursor()
    c.execute("INSERT INTO zoo VALUES (:animal, :number, :price)", {
        'animal': animal.get(),
        'number': number.get(),
        'price': price.get()
        })

    animal.delete(0, END)
    number.delete(0, END)
    price.delete(0, END)


def query():
    conn = sqlite3.connect('zoo.db')
    c = conn.cursor()
    c.execute('SELECT * FROM zoo')
    rows = c.fetchall()

    print_records = ''
    for row in rows:
        print_records += str(row) + "\n"
    
    query_label = ttk.Label(window, text=print_records)
    query_label.grid(row=5, column=0, columnspan=2)

    conn.commit()
    conn.close()
    

animal_label = ttk.Label(window, text='animal').grid(row=0, column=0, padx=20)
number_label = ttk.Label(window, text='number').grid(row=1, column=0, padx=20)
price_label = ttk.Label(window, text='price').grid(row=2, column=0, padx=20)

update_btn = ttk.Button(window, text="Add Data", command=update)
update_btn.grid(row=3, column=0, columnspan=2, pady=10, padx=10, ipadx=100)
query_btn = ttk.Button(window, text='Show Data', command=query)
query_btn.grid(row=4, column=0, columnspan=2, pady=10, padx=10, ipadx=100)

window.mainloop()
```
