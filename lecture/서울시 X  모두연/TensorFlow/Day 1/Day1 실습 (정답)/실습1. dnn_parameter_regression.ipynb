## DNN Function Approximation 신경망 파라미터 변경

#### Package Import

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Activation Function

```python
def activation_1(n):
    return 1/(1+np.exp(-n))

def activation_2(n):
    return n
```

#### 가중치와 편향

```python
p = np.arange(-2, 2, 0.01)
p = p[np.newaxis, : ] # 입력 데이터는 (1, 400) 형태
```

####  첫번째 계층의 편향 b1의 두번째 편향을 변경

```python
W1 = np.array([[10],[10]],float)
b1 = np.array([[-10],[10]],float)
W2 = np.array([1,1],float)
b2 = np.array([[0]], float)

for i in np.linspace(-10, 10, 5):
    b1 = np.array([[-10],[10+i]],float)
    a1 = activation_1(np.dot(W1, p) + b1)
    a2 = activation_2(np.dot(W2, a1) + b2)
    plt.plot(p.flatten(), a2.flatten())
    
plt.axis([-2, 2, -1, 3])
plt.show()
```

#### W2의 첫번째 가중치 변경

```python
W1 = np.array([[10],[10]],float)
b1 = np.array([[-10],[10]],float)
W2 = np.array([1,1],float)
b2 = np.array([[0]], float)

for i in np.linspace(-2,0.0, 5):
    W2 = np.array([1+i,1],float)
    a1 = activation_1(np.dot(W1, p) + b1)
    a2 = activation_2(np.dot(W2, a1) + b2)
    plt.plot(p.flatten(), a2.flatten())
    
plt.axis([-2, 2, -1, 3])
plt.show()
```

#### W2의 두번째 가중치 변경

```python
W1 = np.array([[10],[10]],float)
b1 = np.array([[-10],[10]],float)
W2 = np.array([1,1],float)
b2 = np.array([[0]], float)

for i in np.linspace(-1.5,0.2, 5):
    W2 = np.array([1,1+i],float)
    a1 = activation_1(np.dot(W1, p) + b1)
    a2 = activation_2(np.dot(W2, a1) + b2)
    plt.plot(p.flatten(), a2.flatten())
    
plt.axis([-2, 2, -1, 3])
plt.show()
```

#### 두번째 계층의 편향 b2 변경

```python
W1 = np.array([[10],[10]],float)
b1 = np.array([[-10],[10]],float)
W2 = np.array([1,1],float)
b2 = np.array([[0]], float)

for i in np.linspace(-1,1, 5):
    b2 = np.array([[0+i]], float)
    a1 = activation_1(np.dot(W1, p) + b1)
    a2 = activation_2(np.dot(W2, a1) + b2)
    plt.plot(p.flatten(), a2.flatten())
    
plt.axis([-2, 2, -1, 3])
plt.show()
```
