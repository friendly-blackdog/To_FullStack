## 4) 히스토그램

히스토그램은 수치형 데이터의 분포를 나타낸다. 흔히 빈도(도수), 빈도밀도(도수밀도), 확률, 확률밀도등의 분포를 그릴 때 사용한다. 램에 사용되는 용어는 다음과 같습니다. 
- 가로축 
    - 계급: 변수의 구간, bin (or bucket) 

- 세로축
    - 도수: 빈도수, frequency 

- 전체 총 량: n

### 히스토그램 Remind

잠깐, 히스토그램의 개념 정리한번 하도록 할께요 

우선 히스토그램은 도수분포표를 그래프로 나타낸 것입니다. 

히스토그램의 용어는 다음과 같습니다. 
- 가로축 
    - 계급: 변수의 구간, bin (or bucket) 

- 세로축
    - 도수: 빈도수, frequency 

- 전체 총 량: n

다음과 같은 데이터의 히스토그램을 만들어보겠습니다. 

- x1과 x2의 평균이 100, 130. 
- 도수를 50개의 구간으로 표시하며, 확률 밀도가 아니라 빈도로 표기합니다. 

```python
mu1, mu2, sigma = 100, 130, 15
x1 = mu1 + sigma*np.random.randn(10000)
x2 = mu2 + sigma*np.random.randn(10000)
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
patches = ax1.hist(x1, bins=50, normed=False)
patches = ax1.hist(x2, bins=50, normed=False, alpha=0.5)
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
plt.xlabel('Bins')
plt.ylabel('Number of Values in Bin')
ax1.set_title('Two Frequency Distributions')
plt.show()
```

## 3. 주석 추가

그래프 안에 추가적으로 글자나 화살표등 주석을 그릴때 annotate() 메소드를 이용합니다. 
- ax.annotate()

```python
from datetime import datetime

data = pd.read_csv('./image/AMZN.csv',index_col=0, parse_dates=True)
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
price = data['Close']
price.plot(ax=ax, style='black')

important_data = [(datetime(2019, 6, 3), "Low Price"),(datetime(2020, 2, 19), "Peak Price")]

for d, label in important_data:
    ax.annotate(label, xy=(d, price.asof(d)+10), 
                xytext=(d,price.asof(d)+100),
                arrowprops=dict(facecolor='red'))

plt.grid()
plt.ylim([1600,2200])
plt.xlim(['2019-05-01','2020-03-01'])
ax.set_title('StockPrice')
plt.show()
```
