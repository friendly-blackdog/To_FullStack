```python
#데이터분석 2 : 각 연도별 남녀의 그룹별 / 각 이름의 출생횟수가 전체에서 차지하는 비중 계산
#연도(‘Year’)와 성별(‘Gender’) 의 기준열을 가지고 group하여 딕셔너리 형태로 키값 출력
group_names_dict = dict(list(names.groupby(['Year', 'Gender'])))
group_names_dict.keys()
```

```python
# 샘플로 1990년생 남자(‘M’)에 해당하는 데이터를 추출
sample_1990_M = group_names_dict[(1990, 'M')]
sample_1990_M.head()
```

```python
# 연도(‘Year’)와 성별(‘Gender’) 의 기준열을 가지고 group한 결과물 데이터프레임에 출생비율을 계산하여 새로운열로 
# 추가해줄 사용자 정의함수를 정의하고, 이를 이용하여 비율 계산 후 새로운 'prop'열에 저장
# 출생비율(‘prop’) = [Count 열 변수 / Count 열 변수.sum() 총합]
def Count_prop(df):
    Count = df['Count']
    df['prop'] = Count/Count.sum()
    return df
    
Count_with_prop = names.groupby(['Year', 'Gender']).apply(Count_prop)
Count_with_prop.head()
```
