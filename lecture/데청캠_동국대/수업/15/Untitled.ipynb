```python
data = pd.read_csv("multi_index_ex1.csv", encoding='cp949', sep=',')
```

```python
data.head()
```

```python
data_new = data.set_index(['지역','지역.1']) #index 설정
data_new.index.names=['품목','상세품목'] #index name 설정
```

```python
data_new.columns = [data_new.columns.map(lambda x:x[:2]), 
                    data_new.iloc[0]] # column설정
data_new = data_new.iloc[1:]    # set_column 메서드가 없으므로 원본에서 제거 필요 0행들을 삭제하고 다시 저장
data_new.columns.names=['지역','지점'] # column name 설정
```

```python
data_new = data_new.astype('int') # 데이터를 숫자형 타입으로 변환
data_new
```

## 2) 컴퓨터의 서울지역 판매량 출력 

```python
data_new.xs('컴퓨터').xs('서울', axis=1)
data_new.loc['컴퓨터', '서울']
```

## 3) 서울지역의 컴퓨터의 각 세부항목별 판매량의 합계 출력 

```python
data_new.loc['컴퓨터', '서울'].apply(sum, 1)
```

## 4) 각 지역의 A지점의 TV 판매량 출력 

```python
data_new.xs('A', level=1, axis=1).xs('TV', level=1)
```

## 5) 각 지역 C지점의 모바일의 각 세부항목별 판매량 평균 출력 

```python
round(data_new.xs('C', level=1, axis=1).xs('모바일').mean(1),2)
```

## 6) 서울지역의 A지점의 노트북 판매량 출력 

```python
data_new['서울','A'].xs('노트북', level=1)
```
